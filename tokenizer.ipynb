```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
prompt = "Write an email to customer care to report about my recent nike shoe purchase. Explain what happened.<|assistant|>"
```

```python
# Tokenize the input prompt (encoding)
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
```

```python
print(input_ids)
```

```python
# translate the input_ids [tokens] back to [text] (decoding)
for id in input_ids[0]:
   print(tokenizer.decode(id))
```

```python
# decoding individual tokens to understand how the LLM breaks the words in tokens and later construct back the word.
print(tokenizer.decode(6836)) #ni
print(tokenizer.decode(446)) #ke
print(tokenizer.decode([6836, 446])) #nike
```

```python
# Generate the text using the input tokens
generation_output = model.generate(
  input_ids=input_ids,
  max_new_tokens=20
)
```

```python
# Print the output
print(tokenizer.decode(generation_output[0]))
```
